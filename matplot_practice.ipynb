```python
labels = ['A','B','C']
values  = [1,4,5]
plt.figure(figsize=[10,5], dpi=50)
bars = plt.bar(labels,values , label= 'Bar Values')


#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

# Another Method
pattern = ['/' ,'o' , '*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))
plt.legend()
plt.show()
```

# Scientific Notation  -- Use Matplotlob - Ticker

```python
from matplotlib import ticker
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,y2)
ax.set_title("Scientific Notation")

formatter = ticker.ScalarFormatter(useMathText= True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,2))
ax.yaxis.set_major_formatter(formatter)
```
